## Context
The dataset provided is from Netflix and contains the different detailed information about movies and series that were released on Netflix. The aim is to perform a detailed analysis of this dataset and provide useful information for movies and series that exist on Netflix.

## Data Description:

The detailed data dictionary is given below:

* Title -  The name of the movie/series
* Genre - Movie type
* Tags - Movie type in more detail
* Languages - The language that the movies are available in
* Series or Movie - To determine if the program is a movie or series
* Hidden Gem Score - Scores on Hidden Gem
* Country Availability - The country the movies is available in
* Runtime	- The length of the movie/series
* Director - The director of the movie/series
* Writer - The writer of the movie/series
* Netflix Link	- The link to watch the movie/series on Netflix
* IMDb Link	- The link to view movie/series information on IMDb.
* Summary	- The summary of what the movie/series is about
* IMDb Votes	- The votes on IMDb database
* Image	- The movie cover
* Poster
* TMDb Trailer	- The preview of the movie/series
* Trailer Site - The site to view the preview of the movie/series
